```python
for col in df.columns:
    if col != 'outcome':
        zero_count = (df[col] == 0).sum()
        if zero_count > 0:
            print(f"Column '{col}' has {zero_count} zero values")
```

```python
(df==0).any()
```

```python
# columns where 0 is not a valid medical value

invalid_zero_cols = [
    'glucose',
    'bloodpressure',
    'skinthickness',
    'insulin',
    'bmi'
]

# replace 0 with nan
df[invalid_zero_cols] = df[invalid_zero_cols].replace(0, np.nan)
```

```python
df.sum()
```

```python
df.isnull().sum()
```

```python
#  import missing values
for col in invalid_zero_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# median represents :-> a typical patient not too high not too low not
#  influenced but outliers
```

```python
# feature target sepration
# feature
x=df.drop('outcome',axis=1)
# target
y=df['outcome']
```

```python
# train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
# here using logistic regression classifier because it is
#  a binary classification problem ,storng base line medecal data
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x_train_scaled,y_train)

# here now the model trained ,uses scaled features,can implement safety
```

```python
y_pred=log_reg.predict(x_test_scaled)
```

```python
y_pred[:10]
```

```python
# accurancy of test
log_reg.score(x_test_scaled,y_test)
```

```python
# accurancy of train
log_reg.score(x_train_scaled,y_train)
```

```python
# linear regression probability prediction
y_prob=log_reg.predict_proba(x_test_scaled)
```

```python
 y_prob[:20]
```

```python
y_prob_diabetic = y_prob[:, 1]
y_prob_diabetic[:10]
```

```python
# confusion matrix

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm

# [TN FP]
# [FN TP]

# TN - true negative-actually non dibetic
# FP - false positive-actually non diabetic but predicted diabetic
# FN - false negative-actually diabetic but predicted non diabetic
# TP - true positive-actually diabetic
```

```python
# classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))

# here we will see about precision,recall,f1-score and support

# precision: how correct the model was when  it said 0 or 1
# recall: how many real 0 and 1 it found
# fitbalance: score of precision and recall
# support: how many actual 0 and 1 patients

# example for weighted avg
# classes with more sample influence the final score more than
# the smaller classes
```

```python
# next model we using here is svm support vector
# mechine ,it finds a string ,margin based decision
# boundary that often improve recall on small,
# well scalled datasets

# the main reason to use is datasets is small in size,we did the scalling ,chances are
# high to get better recall
```

```python
from sklearn.svm import SVC
svm_model=SVC(kernel='linear',probability=True)
svm_model.fit(x_train_scaled,y_train)
```

```python
y_pred_svm=svm_model.predict(x_test_scaled)
# evaluating
svm_model.score(x_test_scaled,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_svm=confusion_matrix(y_test,y_pred_svm)
cm_svm
```
